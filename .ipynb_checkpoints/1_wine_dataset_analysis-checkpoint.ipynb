```python
# Импорт необходимых модулей
import pandas as pd
import seaborn as sns
from pylab import rcParams
from matplotlib import pyplot
```

```python
# Импорт данных
red = pd.read_csv('winequality-red.csv', sep =';')
white = pd.read_csv('winequality-white.csv', sep =';')

# просмотр красного 
red.head()
```

```python
# просмотр белого 
white.head()
```

```python
# выведем названия столбцов/физико-химических свойств красного вина
print(red.columns)
```

```python
# просмотр количества пропущенных значений 
white.isnull().sum()
```

```python
# размер массива 
print('Размер массива Red Wine:', red.shape)
print('Размер массива White  Wine:', white.shape)
```

```python
# зададим занчения размера графика
rcParams['figure.figsize'] = 5, 5

# просмотр количества классов для красного вина
print('Качество красного вина от 0 до 10:')
sns.countplot(x='quality', data=red)
```

```python
# просмотр количества классов для красного вина
print('Качество белого вина от 0 до 10:')
sns.countplot(x='quality', data=white)
```

Как видно из приведенных рисунков, классы несбалансированы. Также отсутствуют вина оцененные на 0,1,2,10 для всех и 9 для красного вина.

```python
# Посмотрим сколько именно раз встречается тот или иной класс

print('Красное вино, классы и число их появления:')
print(red['quality'].value_counts(normalize=False))

print('\nБелое вино, классы и число их появления:')
print(white['quality'].value_counts(normalize=False))
```

```python
# посмотрим на корреляцию признаков красного вина 
rcParams['figure.figsize'] = 9, 9
corr = red.corr()
sns.heatmap(corr, cmap="bwr",annot=True, linewidths=.5);
```

Как вывод по корреляции: 
Данные слабо коррелированы, необходимость сокращения количества признаков, за счет коррелирующих данных, отсутствует

```python
# посмотрим на корреляцию признаков белого вина 
corr = white.corr()
sns.heatmap(corr, cmap="bwr",annot=True, linewidths=.5);
```

Почти аналогичный вывод за исключением положительной корреляции между residual sugar (остаточный сахар) и плотности. Объясняется тем, что наличие сахара (не перебродившего в алкоголь) в жидкости делает её более плотной, густой. 
Ровно тем же самым объясняется отрицательная корреляция между alcohol и residual sugar, а также между alcohol и density, т.к. чем выше содержание алкоголя, тем меньше остаточного сахара (весь сахар перебродил), а как следствие пониженная плотность жидкости.
Такая же зависимость между индексом pH и кислотностью.

```python
# посмотрим гистограмму распределения признаков красного вина

rcParams['figure.figsize'] = 12, 12

red_features = red.drop(columns = ['quality'])

print('Красное вино диаграмма распределения:')
red_features.hist()               
pyplot.show()
```
